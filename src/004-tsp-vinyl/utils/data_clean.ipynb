Pedi ao ChatGPT que montasse uma planilha com algumas informações de lojas de discos de vinil da cidade de São Paulo e depois, na primeira linha, eu adicionei as informações de um hotel.

Neste notebook, eu fiz mais algumas transformações nessa base para poder usá-la de forma a resolver o problema do caixeiro viajante aplicado a visitas a lojas de vinil de SP.

### Importando as bibliotecas

```python
import pandas as pd
import time
from geopy.geocoders import Nominatim
```

### Carregando a base de dados criada pelo ChatGPT

```python
df = pd.read_csv('../data/data.csv')
df.head()
```

### Imprimindo as dimensões desse dataframe

```python
df.shape
```

### Criando as colunas latitude e longitude

Usando a biblioteca geopy

```python
geolocator = Nominatim(user_agent='test_app')
```

Abaixo são dois exemplos da sua utilização

```python
location = geolocator.geocode('Rua dos Trilhos, 1212 - Mooca, São Paulo - SP')
```

```python
location = geolocator.geocode('Rua Quintino Bocaiúva, 285 - Sé, São Paulo - SP')
location
```

Obtendo a latitude e longitude para cada localização presente no dataframe e adicionando as novas colunas

```python
lat = []
lon = []

for i, row in df.iterrows():
    try:
        location = geolocator.geocode(row['endereço'])
        if not location:
            location = geolocator.geocode(row['cep'])
        print(location)
        lat.append(location.latitude)
        lon.append(location.longitude)
    except:
        lat.append(None)
        lon.append(None)
    time.sleep(10)
df['latitude'] = lat
df['longitude'] = lon
```

Imprimindo o dataframe atualizado

```python
df
```

Salvando o dataframe atualizado para ser utilizado para resolver o problema do caixeiro viajante

```python
df.to_csv('../data/data_t.csv', index=False)
```
